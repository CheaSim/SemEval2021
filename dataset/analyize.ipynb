```python

import seaborn as sns
colors = sns.color_palette(n_colors=4)
def show_length(file_path):
    if file_path == None:
        file_path = './task3/3.jsonl'
    t_i = 0
    for i in range(10):
        if str(i) in file_path[7:]:
            t_i = i
    y = []
    with open(file_path, 'r', encoding='UTF-8') as file:
        for line in file.readlines():
            t = json.loads(line) # t single json file = article question answers
            y.append(len(t['article'].split(' ')))
    y.sort()
    x = [1.0*i/len(y) for i in range(len(y))]
    plt.scatter(x,y, label=t_i , color=colors[t_i], s=1, marker="o")
    plt.xlabel('sample')
    plt.ylabel('length of article')
    plt.title('article length')
    plt.legend()
    # plt.show()
    plt.subplot(111)
    return x,y
a,b = show_length('./task3/1.jsonl')
a1,b1 = show_length('./task3/2.jsonl')
a2,b2 = show_length('./task3/3.jsonl')

plt.figure(111)
plt.subplot(111)
plt.plot(a,b)
plt.figure()
plt.subplot(121)
plt.plot(a1,b1)
plt.subplot(131)
plt.plot(a2,b2)
```
